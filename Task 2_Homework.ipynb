# Handling Missing Values

```python
missing_values = crime_data.isnull().sum()
missing_values
```

```python
# Fill missing values for categorical columns with "Unknown"
crime_data.fillna({
    'Vict Sex': "Unknown",
    'Vict Descent': "Unknown",
    'Premis Desc': "Unknown",
    'Weapon Desc': "Unknown",
    'Status': "Unknown",
    'Cross Street': "Unknown",
    'Mocodes': "Unknown"
}, inplace=True)

# Fill missing values for numerical columns with 0
crime_data.fillna({
    'Premis Cd': 0,
    'Weapon Used Cd': 0,
    'Crm Cd 1': 0,
    'Crm Cd 2': 0,
    'Crm Cd 3': 0,
    'Crm Cd 4': 0
}, inplace=True)
```

```python
missing_values = crime_data.isnull().sum()
missing_values
```

# Data Cleaning and Transformation

```python
# Standardize column names
crime_data.columns = crime_data.columns.str.lower().str.replace(' ', '_')
```

```python
crime_data['date_occ'] = pd.to_datetime(crime_data['date_occ'], format='%m/%d/%Y', errors='coerce')  # Change format as necessary

# Extract year, month, day from date_occ
crime_data['year'] = crime_data['date_occ'].dt.year
crime_data['month'] = crime_data['date_occ'].dt.month
crime_data['day'] = crime_data['date_occ'].dt.day

# Extract hour from 'time_occ' (HHMM format)
crime_data['hour'] = (crime_data['time_occ'] // 100).astype(int)

# Create is_weekend column (0 = Monday, 6 = Sunday)
crime_data['is_weekend'] = crime_data['date_occ'].dt.dayofweek >= 5  # Saturday and Sunday
```
